```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get("https://internshala.com/jobs/data-science-jobs/page-1/",headers = headers).text

soup = BeautifulSoup(webpage, 'lxml')
com = soup.find_all("div",class_= "individual_internship")

# Company Names
name_list = []

# Assuming 'com' is a list of all internship divs
for internship in com:  # Iterate over each internship div in 'com'
    location_div = internship.find('p', class_='company-name')  # Find the location div

    # Check if the location div is found and extract the text
    if location_div:
        location = location_div.text.strip()  # Extract the text from the <a> tag inside the location div
        name_list.append(location)
    else:
      name_list.append("No data")

location_list = []
# Assuming 'com' is a list of all internship divs
for internship in com:  # Iterate over each internship div in 'com'
    location_div = internship.find('p', class_='locations')  # Find the location div

    # Check if the location div is found and extract the text
    if location_div:
        location = location_div.find('a').text.strip()  # Extract the text from the <a> tag inside the location div
        location_list.append(location)
    else:
      location_list.append("No data")

experience_list = []
for internship in com:
    experience_div = internship.find('div', class_='row-1-item')

    # Check if the experience div is found and extract the text
    if experience_div:
        experience = experience_div.text.strip()
        experience_list.append(experience)
    else:
      experience_list.append("No data")

type_list = []
for internship in com:
    status_div = internship.find('div', class_='status-li')

    # Check if the experience div is found and extract the text
    if status_div:
      experience = status_div.text.strip()
      type_list.append(experience)
    else:
      type_list.append("No data")

title_list = []
for internship in com:
    status_div = internship.find('h3', class_='job-internship-name')

    # Check if the experience div is found and extract the text
    if status_div:
      experience = status_div.text.strip()
      title_list.append(experience)
    else:
      title_list.append("No data")

salary_list = []
for salary in com:
    salary_div = salary.find('span', class_='desktop')  # The salary info is usually in a 'span' with class 'desktop'

    # Check if the salary div is found and extract the text
    if salary_div:
        salary = salary_div.text.strip()
        salary_list.append(salary)
    else:
      salary_list.append("No data")

df = pd.DataFrame({"Title": title_list,
              "Type":type_list,
              "Company":name_list,
              "CTC":salary_list,
              "Location":location_list,
              "Exp":experience_list})
print(df)
```
